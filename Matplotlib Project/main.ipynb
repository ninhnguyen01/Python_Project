```python
# Import package
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Read file
weather = pd.read_csv('/Matplotlib Project/seattle_weather.csv')
# Preview first few rows
weather.head()
```

```python
# View columns
weather.columns.unique()
```

```python
# Rename column
weather = weather.rename(columns={'DATE':'MONTH'})
```

```python
# Preview first few rows
weather.head()
```

```python
# Line Graph
sns.lineplot(x=weather["MONTH"],y=weather["MLY-PRCP-NORMAL"], marker="o") 
plt.show() 
```

```python
# Read the file
medals = pd.read_csv('/Matplotlib Project/summer2016.csv', index_col=0)  
# Preview first few rows
medals.head()
```

```python
# Bar graph of gold medal by sport
sport_medal = sns.barplot(medals, y='Sport', x=medals['Medal'] == 'Gold', hue='Sport')
plt.xlabel('Gold Medal by Percentage')
plt.figure(figsize=[10,8])
plt.show()
```
