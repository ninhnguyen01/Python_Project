```python
# Import package
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read file
admissions = pd.read_csv('/Admission Project/admissions.csv')
# Create pairplots 
pairplot = sns.pairplot(admissions[['Serial No.','GRE Score', 'Chance of Admit']])
plt.show()
plt.close()
```

```python
# Print Pearson correlation coefficient 
print(admissions['GRE Score'].corr(admissions['Chance of Admit']))
# Plot correlations heatmap
heatmap = sns.heatmap(admissions.corr(),annot=True) 
plt.show()
plt.close() 
```
