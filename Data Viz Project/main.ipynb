# Examining Pollution Visualization Project

```python
# Import package
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read file
pollution = pd.read_csv('/Data Viz Project/pollution.csv')
# Preview first few rows
pollution.head()
```

```python
# View all unique cities
pollution.city.unique()
```

```python
pollution_long_beach = pollution[pollution.city == 'Long Beach']
# Make an array of colors based upon if a row is a given day 
long_beach_colors = ['orangered' if day == 38 else 'steelblue' for day in pollution_long_beach.day]
# Plot with additional scatter plot argument facecolors  
p = sns.regplot(x='NO2', y='SO2', data = pollution_long_beach, fit_reg=False, scatter_kws={'facecolors': long_beach_colors,'alpha': 0.7}) 
plt.title('Pollution in Long Beach')
plt.show()
```

```python
# Scatterplot
# Assign a new column to dataframe the desired combos 
pollution['interesting cities'] = [x if x in ['Long Beach', 'Cincinnati'] else 'other' for x in pollution['city'] ]  
sns.scatterplot(x="NO2", y="SO2", hue = 'interesting cities', palette='Set2', data=pollution.query('year == 2014 & month == 12'))
plt.show()
```
