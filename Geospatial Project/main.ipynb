# Geospatial Project

```python
# Import package
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
# Read file
school = gpd.read_file('/Geospatial Project/schools.csv')
# Preview first few rows
school.head()
```

```python
# View data type
school.info()
```

```python
# Change data types
school['School ID'] = school['School ID'].astype('int64')
school['School Name'] = school['School Name'].astype('string')
school['Street Address'] = school['Street Address'].astype('string')
school['City'] = school['City'].astype('string')
school['State'] = school['State'].astype('string')
school['ZIP Code'] = school['ZIP Code'].astype('int64')
school['Phone Number'] = school['Phone Number'].astype('string')
school['School State ID'] = school['School State ID'].astype('int64')
school['School Level'] = school['School Level'].astype('string')
school['Lowest Grade'] = school['Lowest Grade'].astype('string')
school['Highest Grade'] = school['Highest Grade'].astype('string')
school['Latitude'] = school['Latitude'].astype('float64')
school['Longitude'] = school['Longitude'].astype('float64')
school['School Website'] = school['School Website'].astype('string')
school['Cluster'] = school['Cluster'].astype('string')
school['Mapped Location'] = school['Mapped Location'].astype('string')
```

```python
# View new data types
school.info()
```
