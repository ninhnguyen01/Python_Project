# Homelessness Project

```python
# Import package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read file
homeless = pd.read_csv('/Data Manipulation Project/homelessness.csv')
# Preview first few rows
homeless.head()
```

```python
# Drop column
homeless = homeless.drop(columns=['Unnamed: 0'])
# Preview first few rows
homeless.head()
```

```python
# View data
homeless.info()
```

```python
# Change data type
homeless['region'] = homeless['region'].astype('category')
homeless['state'] = homeless['state'].astype('category')
```

```python
# View new data
homeless.info()
```

```python
# Sort
homeless.sort_values('state')
```

```python
# Subset
homeless[['state','individuals']]
```

```python
# Mean
homeless[['individuals','family_members']].mean()
```

```python
# Count by region
homeless['region'].value_counts()
```

```python
# Count by region with normalized data
homeless['region'].value_counts(normalize=True)
```

```python
# Mean Summaries by region and individual
print(homeless[homeless['region'] == 'South Atlantic']['individuals'].mean())
print(homeless[homeless['region'] == 'Mountain']['individuals'].mean())
print(homeless[homeless['region'] == 'West North Central']['individuals'].mean())

print(homeless[homeless['region'] == 'New England']['individuals'].mean())
print(homeless[homeless['region'] == 'East North Centrall']['individuals'].mean())
print(homeless[homeless['region'] == 'Pacific']['individuals'].mean())

print(homeless[homeless['region'] == 'East South Central']['individuals'].mean())
print(homeless[homeless['region'] == 'West South Central']['individuals'].mean())
print(homeless[homeless['region'] == 'Mid-Atlantic ']['individuals'].mean())
```

```python
# Scatter plot
homeless.plot(x="region", y="individuals", kind="scatter")
plt.xticks(rotation=45)
plt.show()
```
