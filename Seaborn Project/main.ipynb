# Data Viz Project

```python
# Import package
import seaborn as sns  
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Read file
people = pd.read_csv('/Seaborn/Seaborn Project/young-people-survey-responses.csv', index_col=0)
# Preview the first few rows
people.head()
```

```python
# Count plot
sns.countplot(data=people, x='Age', hue='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count plot
sns.countplot(data = people, x='Village - town', hue='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count plot
sns.countplot(data = people, x='Internet usage', hue='Gender')
plt.xticks(rotation=45)
plt.show()
```
